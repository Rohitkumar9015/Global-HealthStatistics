# Global Health Statistics

#### This dataset provides detailed global health information with one million entries. It covers diseases, their prevalence, incidence, and mortality rates across various countries and years. It includes data about affected age groups, genders, and healthcare systems, like doctors and hospital beds per 1,000 people. It also highlights recovery rates, treatment types, and vaccine availability. Economic and social factors, such as treatment costs, income, education levels, and urbanization, are included. The dataset measures disease impacts using DALYs and tracks five-year improvements. This data helps researchers, policymakers, and healthcare experts analyze trends and improve health systems globally.


#### The dataset contains the following columns:

##### 1.Country: Name of the country.
##### 2.Year: Year of the data.
##### 3.Disease Name: Specific disease name.
##### 4.Disease Category: Broader category of the disease (e.g., Respiratory, Parasitic).
##### 5.Prevalence Rate (%): Percentage of population affected by the disease.
##### 6.Incidence Rate (%): Percentage of new cases reported annually.
##### 7.Mortality Rate (%): Percentage of deaths caused by the disease.
##### 8.Age Group: Age range of affected individuals.
##### 9.Gender: Gender of affected individuals.
##### 10.Population Affected: Number of people affected by the disease.
##### 11.Healthcare Access (%): The percentage of the population with access to healthcare.
##### 12.Doctors per 1000: The number of doctors per 1000 people
##### 13.Hospital Beds per 1000: Availability of hospital beds per 1,000 population.
##### 14.Treatment Type: Type of treatment (e.g., Medication, Surgery, Vaccination).
##### 15.Average Treatment Cost (USD): Average cost of treatment.
##### 16.Availability of Vaccines/Treatment: Whether vaccines or treatments are available (Yes/No).
##### 17.Recovery Rate (%): Percentage of patients recovering from the disease.
##### 18.DALYs (Disability-Adjusted Life Years): Years of healthy life lost due to disease.
##### 19.Improvement in 5 Years (%): Improvement in health metrics over five years.
##### 20.Per Capita Income (USD): Average income per person in USD.
##### 21.Education Index: A measure of education quality and accessibility.
##### 22.Urbanization Rate (%): Percentage of urban population.

## The code selects specific healthcare-related columns from the DataFrame (df). These columns focus on disease prevalence, mortality, healthcare access, economic factors, and urbanization metrics for analysis.

```python
# Define custom age group categorization
def categorize_age_group(age_group):
    if '-' in age_group:  # Range type age group
        start_age = int(age_group.split('-')[0])
        end_age = int(age_group.split('-')[1])
    else:  # Single age group
        start_age = int(age_group.split('+')[0])
        end_age = start_age
    
    if end_age <= 18:
        return '0-18'
    elif end_age <= 36:
        return '19-36'
    elif end_age <= 54:
        return '37-54'
    elif end_age <= 72:
        return '55-72'
    elif end_age <= 90:
        return '73-90'
    elif end_age <= 108:
        return '91-108'
    elif end_age <= 126:
        return '109-126'
    elif end_age <= 144:
        return '127-144'
    elif end_age <= 162:
        return '145-162'
    elif end_age <= 180:
        return '163-180'
    else:
        return '180'

# Apply the categorization function to create new 'Age Group' column
df['Age Group Categorized'] = df['Age Group'].apply(categorize_age_group)

# Count the number of occurrences for each categorized age group
age_group_counts = df['Age Group Categorized'].value_counts().sort_index(ascending=True)

# Print the resulting Series for graph representation
print(age_group_counts)
```

# Inference:
## The code categorizes age groups into predefined ranges, counts occurrences for each group, and outputs the frequency of individuals in each categorized age range.

```python
## Which countries have seen the most improvement in healthcare outcomes over 5 years? (Bar Chart)

# Strip any possible spaces in column names (in case they exist)
df.columns = df.columns.str.strip()

# Group data to eliminate duplicates
df = df.groupby(['Country', 'Year'], as_index=False).agg({
    'Mortality Rate (%)': 'mean',
    'Recovery Rate (%)': 'mean'
})

# Debugging: Confirm the new unique data
print(df)

# Pivot data to compute the change over 5 years for Mortality Rate
df_pivot = df.pivot(index='Country', columns='Year', values='Mortality Rate (%)')

# Calculate improvement: change from 2018 to 2023
df_pivot['Improvement'] = df_pivot[2023] - df_pivot[2018]

# Handle any NaN values (remove if no valid data to compute)
df_pivot.dropna(subset=['Improvement'], inplace=True)

# Get the top 10 countries with the most improvement
top_countries = df_pivot['Improvement'].nlargest(10).reset_index()

# Plotting the bar chart
plt.figure(figsize=(12, 8))
sns.barplot(
    data=top_countries,
    x='Country',
    y='Improvement',
    palette='viridis'
)

# Customize the plot
plt.title("Top 10 Countries with the Most Improvement in Mortality Rates (Over 5 Years)", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Mortality Rate (%)", fontsize=12)
plt.xticks(rotation=45)  # Rotate for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

# Inference:
## The code identifies and visualizes the top 10 countries with the greatest reduction in mortality rates between 2018 and 2023, showing improvements in healthcare outcomes over 5 years.

```python
## How have the top 5 diseases' DALYs varied over time in the top 10 countries with the highest DALYs?
# Get the top 5 disease categories based on the mean DALYs
top_disease_categories = (df.groupby('Disease Category')['DALYs'].mean().nlargest(5).index)

# Filter the DataFrame to include only the top 5 disease categories
df_top_diseases = df[df['Disease Category'].isin(top_disease_categories)]

# Get the top 3 countries based on the mean DALYs
top_countries = (df.groupby('Country')['DALYs'].mean().nlargest(10).index)

# Filter the DataFrame to include only the top 3 countries
df_top = df_top_diseases[df_top_diseases['Country'].isin(top_countries)]

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_top, x='Year', y='DALYs', hue='Disease Category', marker='o')

# Customize the plot
plt.title('DALYs Over Time for Top Diseases and Countries', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('DALYs', fontsize=14)
plt.legend(title='Disease Category', bbox_to_anchor=(1.05, 1), loc='upper left')

# Make sure everything fits without overlap
plt.tight_layout()

# Show the plot
plt.show()
```
